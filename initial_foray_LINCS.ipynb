```python
# can I make a dictionary between the gene id and gene symbol?
idToSymbol = zip(gene_names.gene_id, gene_names.gene_symbol) # zip creates a dictionary between gene id and the gene symbol
idToSymbol = list(idToSymbol)
idToSymbol = dict(idToSymbol)
# could have just iterated over it every single time but this is efficient!
```

Dictionary Conversion from Entrez to Gene Symbol

```python
converted_first_module_UP = []
int_first_module = list(map(int, first_module)) # fastest way to convert string to integer
for i in int_first_module:
    for key in idToSymbol:
        if i == key:
            converted_first_module_UP.append(idToSymbol[key])

print(len(converted_first_module_UP))
```
